# 1. For implementing PCA, the number of components were tested from 2 to 100, as the n_components are increased, our dataset is able to capture high variance!

# 2. Below function, depicts the variation of n_components and the variation captured


```python
def pca(n):
    pca = PCA(n_components=n)
    pc=pca.fit_transform(t)
    print(pca.explained_variance_)
    print(pca.explained_variance_ratio_)
    print (pca.explained_variance_ratio_.cumsum())
```

```python
pca(2)
##only 32% variation 
```

```python
pca(10)
##only 60% variation
```

```python
pca(20)
## only 71% variation
```

```python
pca(50)
## 84% variation
```

# Out of all the values, when n_component was 100, it's able to capture 94% variation of the data, which being the optimal and furthur used for visualisation

```python
pca(100)
```

```python
names=[]
for i in range(1,101):
    p='PCA' + str(int(i))
    names.append(p)
```
